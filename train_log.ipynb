```python
list_A = ['France', 'Germany', 'Italy', 'Spain', 'United Kingdom', 'United States', 'China', 'Japan', 'India', 'Brazil']
input = input("Enter a country origin: ")
output = input("Enter a country destination: ")



```

```python
for country in list_A:
    if country == input:
        print(f"The origin of the train is {country}")
    else:
        print(f"The origin of the train is not {country}")
for country in list_A:    
    if country == output:
        print(f"The destination of the train is {country}")
    else:
        print(f"The destination of the train is not {country}")
```

```python
while True:
    if input in list_A and output in list_A:
        print("The train can travel from the origin to the destination.")
        print(f"The train departs from {input} and travels through {list_A[list_A.index(input)+1:list_A.index(output)]} before arriving at {output}.")
        break
```
